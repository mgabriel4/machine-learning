```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.svm import SVR
from sklearn.compose import TransformedTargetRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

df = pd.read_csv("/home/mgabriel4/Documentos/GitHub/machine-learning/data/AmesHousing.csv")
```

```python
data = df.drop(columns=["Order", "PID"]) #tirando colunas inúteis

y = data["SalePrice"]
X = data.drop(columns=["SalePrice"])
```

Pré-processamento dos dados.

```python
X_dum = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_dum, y, test_size=0.3, random_state=42
)
```

```python
pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("svr", SVR(kernel="rbf", C=100.0, epsilon=0.1))
])
```

Modelo:

```python
model = TransformedTargetRegressor(
    regressor=pipe,
    transformer=StandardScaler()
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

Coeficientes:

```python
mse = mean_squared_error(y_test, y_pred)
rmse = mse**0.5
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE :", mse)
print("RMSE:", rmse)
print("MAE :", mae)
print("R²  :", r2)
```
